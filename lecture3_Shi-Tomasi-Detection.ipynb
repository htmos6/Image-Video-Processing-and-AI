```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
real_chess_board = cv2.imread('../DATA/real_chessboard.jpg')
real_chess_board = cv2.cvtColor(real_chess_board, cv2.COLOR_BGR2RGB)

flat_chess_board = cv2.imread('../DATA/flat_chessboard.png')
flat_chess_board = cv2.cvtColor(flat_chess_board, cv2.COLOR_BGR2RGB)

gray_real_board = cv2.cvtColor(real_chess_board, cv2.COLOR_BGR2GRAY)
gray_flat_board = cv2.cvtColor(flat_chess_board, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_board, 13, 0.01, 10) # 10 --> minDistance / Minimum possible Euclidean distance between the returned corners. 
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess_board, (x, y), 3, (0, 255, 0), -1)
    
plt.imshow(flat_chess_board)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_board, 25, 0.01, 4) 
corners = np.int0(corners) # Center of the circle needs a integer values. So convert them to integer

for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess_board, (x, y), 3, (0, 255, 0), -1)

plt.imshow(real_chess_board)
```

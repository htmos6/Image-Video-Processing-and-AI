```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    cv2.putText(blank_img,text='ABCDE',org=(50,300), fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5), np.float32)
erosion1 = cv2.erode(img, kernel, iterations = 1)
display_img(erosion1)
```

```python
img = load_img()
erosion1 = cv2.erode(img, kernel, iterations = 4)
display_img(erosion1)
```

```python
white_noise = np.random.randint(0,2,(600,600))
display_img(white_noise)
```

```python
img = load_img()
img.max()
```

```python
white_noise = white_noise * 255
white_noise.max()
```

```python
img = white_noise + img
display_img(img)
```

```python
opening_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
display_img(opening_img)
```

```python
img = load_img()
black_noise = np.random.randint(0,2,(600,600)) * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_img(closing)
```

```python
img = load_img()
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel) # difference between erosion & dialation
display_img(gradient)
```

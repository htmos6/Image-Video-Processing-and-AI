```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread("../DATA/sudoku.jpg", 0)
```

```python
display_image(img)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 0, 1, 7)
sobely = cv2.Sobel(img, cv2.CV_64F, 1, 0, 7)
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
display_image(sobelx)
```

```python
display_image(sobely)
```

```python
display_image(laplacian) # Laplacian provides both x & y edges. However, edges are not shown exactly
```

```python
blended = cv2.addWeighted(src1 = sobely, alpha = 0.5, src2 = sobelx, beta = 0.5, gamma = 10)
display_image(blended)
```

```python
kernel = np.ones((4,4), np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_image(gradient)
```

```python
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
display_image(gradient)
```

```python
used_thresh, th1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
display_image(th1)
```

```python
used_thresh, th2 = cv2.threshold(gradient, 100, 255, cv2.THRESH_BINARY_INV)
display_image(th2)
```

```python
ret,th3 = cv2.threshold(blended,100,255,cv2.THRESH_BINARY_INV)
display_image(th3)
```

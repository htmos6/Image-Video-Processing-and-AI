# Drawing on Images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
blank_image = np.zeros(shape = (512,512,3), dtype = np.int16)
plt.imshow(blank_image)
```

# Rectangle/ Square/ Line/ Circle

```python
cv2.rectangle(blank_image, (300,10), (400, 90), color = (255,0,0), thickness = 5)
plt.imshow(blank_image)
```

```python
cv2.circle(blank_image, center = (400,200), radius = 50, color = (0,0,255), thickness = 7)
plt.imshow(blank_image)
```

# Filled Geometrical shape --> thickness = -1

```python
cv2.circle(blank_image, center = (100,400), radius = 70, color = (0,255,255), thickness = -1)
plt.imshow(blank_image)
```

```python
cv2.line(blank_image, pt1 = (0,0), pt2 = (512,512), color = (0,255,0), thickness = 6)
plt.imshow(blank_image)
```

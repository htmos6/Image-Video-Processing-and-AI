```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
dog_face = cv2.imread('../DATA/sammy_face.jpg')
dog_face = cv2.cvtColor(dog_face, cv2.COLOR_BGR2RGB)
plt.imshow(dog_face)
```

```python
edges = cv2.Canny(dog_face, 127, 127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(dog_face, 0, 255)
plt.imshow(edges)
```

```python
median_val = np.median(dog_face)
print(median_val)

lower = max(0, median_val*0.7)
upper = min(255, median_val*1.3)
edges = cv2.Canny(dog_face, lower, upper+100) # add +100 to obtain better result
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(dog_face, (7,7))
edges = cv2.Canny(blurred_img, lower, upper+ 30) # add +100 to obtain better result
plt.imshow(edges)
```

```python
img = cv2.imread('../DATA/sammy.jpg')
blurr = cv2.blur(img, (9,9))
edges = cv2.Canny(blurr, lower, upper+ 30) # add +100 to obtain better result
display(edges)
```

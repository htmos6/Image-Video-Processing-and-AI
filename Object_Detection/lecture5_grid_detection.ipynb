```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
found, corners = cv2.findChessboardCorners(flat_chess, (7,7)) # since board is 8x8 --> it cannot find 8th squares. So find 1 smaller size 7x7
print(found) # Found is used in real life examples. If found then you can continue to your operations. 
cv2.drawChessboardCorners(flat_chess, (7,7), corners, found)
plt.imshow(flat_chess)
```

```python
dot_img = cv2.imread('../DATA/dot_grid.png')

plt.subplot(121)
plt.imshow(dot_img)

found, corners = cv2.findCirclesGrid(dot_img, (10, 10), cv2.CALIB_CB_SYMMETRIC_GRID)
cv2.drawChessboardCorners(dot_img, (10, 10), corners, found)

plt.subplot(122)
plt.imshow(dot_img)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
sep_coins = cv2.imread('../DATA/pennies.jpg')
display(sep_coins)
```

```python
# Median Blur --> To eliminate unnecessary edges corners etc.
# Gray Scale
# Binary Threshold --> For black and white image. To seperate background and forground objects. 
# Find Contours
```

```python
median_blurred = cv2.medianBlur(sep_coins, 25)
gray_scale = cv2.cvtColor(median_blurred, cv2.COLOR_BGR2GRAY)
ret, th1 = cv2.threshold(gray_scale, 160, 255, cv2.THRESH_BINARY)
ret2, th2 = cv2.threshold(gray_scale, 160, 255, cv2.THRESH_BINARY_INV)

display(th1)
display(th2)
```

```python
# Lets find contours

contours, hierarchy = cv2.findContours(th2.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coins, contours, i, (0, 255, 0), 10)
display(sep_coins)
```

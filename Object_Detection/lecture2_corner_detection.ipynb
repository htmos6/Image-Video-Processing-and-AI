```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess_board = cv2.imread("../DATA/real_chessboard.jpg")
real_chess_board = cv2.cvtColor(real_chess_board, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess_board)
```

```python
gray_real_chess = cv2.cvtColor(real_chess_board, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)
dest = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k = 0.04)
dest = cv2.dilate(dest, None) # It highlightes corners very clearly  # Not affects the overall result
flat_chess[dest > 0.01*dest.max()] = [255, 0, 0]
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
dest = cv2.cornerHarris(gray, 2, 3, 0.04)
dest = cv2.dilate(dest, None) # Not affects the overall result
real_chess_board[dest > 0.01*dest.max()] = [0, 255, 0]
plt.imshow(real_chess_board)
```
